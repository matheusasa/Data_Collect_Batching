```python
# Inicialização do Ambiente do Cluster (SparkSession)
from pyspark.sql import SparkSession

print("Iniciando a SparkSession de teste...")

# Instancia a aplicacao basica e ativa a JVM (Java Virtual Machine) interna
spark = (SparkSession.builder
    .appName("Teste-Conectividade-Spark")
    .getOrCreate()
)

print(f"SparkSession ativa com sucesso. Versao do Apache Spark: {spark.version}")
```

```python
# Validação de Processamento de Dados em Memória

print("Testando motor de execucao distribuido na memoria RAM...")

try:
    # Cria uma lista de tuplas estruturadas diretamente no script
    dados_teste = [("Postgres", "Online"), ("MinIO", "Online"), ("Spark", "Online")]
    colunas_teste = ["Servico", "Status_Verificado"]
    
    # Transforma a lista local em um Spark DataFrame estruturado
    df_verificacao = spark.createDataFrame(dados_teste, colunas_teste)
    
    # Forca a acao de impressao em tela para validar o pipeline
    df_verificacao.show()
    print("Processamento e montagem de estruturas concluidos com sucesso!")

except Exception as e:
    print(f"Falha critica: Erro durante o processamento de dados no Spark. Erro: {e}")
```

```python
# Desalocação e Encerramento da Sessão
print("Encerrando a SparkSession e limpando os recursos do cluster...")
spark.stop()
print("Sessao desativada com seguranca.")
```

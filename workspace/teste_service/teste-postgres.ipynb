```python
# Instalação de Dependências e Configuração
%pip install -q psycopg2-binary

import psycopg2

# Dicionario com as credenciais de acesso ao banco transacional
config_db = {
    "host": "postgres", 
    "database": "postgres", 
    "user": "postgres", 
    "password": "tailwind2026", 
    "port": 5442
}
```

```python
# Teste de Conexão e Execução de Consulta

print("Iniciando teste de conectividade com o PostgreSQL...")

try:
    # Tenta abrir o socket de conexao com o servidor
    conn = psycopg2.connect(**config_db)
    print("Conexao estabelecida com sucesso!")
    
    # Abre um cursor para enviar comandos SQL
    with conn.cursor() as cursor:
        # Executa uma funcao interna do Postgres para retornar a versao ativa
        cursor.execute("SELECT version();")
        versao_banco = cursor.fetchone()
        print(f"Confirmacao do Servidor: {versao_banco[0]}")
        
    # Fecha a conexao explicitamente para liberar recursos do banco
    conn.close()
    print("Conexao encerrada corretamente. Recursos liberados.")

except Exception as e:
    print(f"Falha critica: Nao foi possivel conectar ao PostgreSQL. Erro: {e}")
```

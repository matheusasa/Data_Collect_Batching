```python
# Instalação de Dependências e Configuração
%pip install -q boto3

import boto3

# Mapeamento do ponto de acesso HTTP local e credenciais do MinIO
config_s3 = {
    "endpoint_url": "http://minio:9000", 
    "aws_access_key_id": "root-minio", 
    "aws_secret_access_key": "root12345678", 
    "region_name": "local"
}
```

```python
# Teste de Autenticação e Listagem de Buckets
print("Iniciando teste de conectividade com o MinIO...")

try:
    # Inicializa o cliente HTTP estruturado do Boto3
    s3_client = boto3.client('s3', **config_s3)
    
    # Dispara uma chamada de API para listar as pastas raiz (buckets)
    resposta = s3_client.list_buckets()
    print("Conexao e autenticacao realizadas com sucesso!")
    
    # Varre a resposta estruturada para exibir os buckets existentes
    print("Buckets localizados no Data Lake:")
    for bucket in resposta.get('Buckets', []):
        print(f" -> Nome do Bucket: {bucket['Name']}")

except Exception as e:
    print(f"Falha critica: Nao foi possivel estabelecer comunicacao com o MinIO. Erro: {e}")
```
